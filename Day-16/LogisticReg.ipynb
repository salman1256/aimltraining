## 2. Logistic Regression — Pass/Fail

```python
#imported all required lib
import pandas as pd
import numpy as np
from sklearn.linear_model import  LogisticRegression
```

```python
#collect the data
df=pd.read_csv('pass_fail.csv')
df.head()
```

```python
#spilt the data
X = df[['Hours']]
y = df['Result']


```

```python
#create the model (change here only we are using LogisticRegression)
model=LogisticRegression()
```

```python
#trained the Model
model.fit(X,y)
```

```python
pred_35 = model.predict([[3.5]])
print('Predicted result for 3.5 hours:', pred_35[0])
```

```python
hours_study=float(input('Number of Hours studied'))
pred_value=model.predict([[hours_study]])
print(f'After {hours_study} hours study you are likely to {pred_value[0]}')
```

```python
#imported all required lib
from sklearn.preprocessing import LabelEncoder 
```

```python
# collect the data
df=pd.read_csv('customer_purchase.csv')
df
```

```python
#data spilt and labelling
le=LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])
X = df[['Age', 'Salary']]
y = df['Purchased']
df
```

```python
#creating the model
model=LogisticRegression()
```

```python
#trained the model
model.fit(X,y)
```

```python
#predict the result
new_customer = [[28, 35000]]
prediction = model.predict(new_customer)
print(f"Predicted Result (1=Yes, 0=No): {prediction[0]}")
```

```python
#prediction: predict the result
new_cust_age=int(input('Enter Age'))
new_cust_sal=int(input('Enter Salary'))
new_cust = [[new_cust_age, new_cust_sal]]
prediction = model.predict(new_cust)
if (prediction):
    print('likely Purchased')
else:
    print('likely Not Purchased')
```

# Logistic Regression — Spam/NotSpam
## Download spam_data.csv from https://www.kaggle.com/
## Load the Data
## Do the data cleaning and labeling (if required)
## Use Logistic Regression Algorithm to trained the Model
## Predict Result if the email is spam or not spam


```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
from sklearn.linear_model import  LogisticRegression
```

```python
df = pd.read_csv("spam_dataset.csv")   
df
```

```python
X = df['message_content']
y = df['is_spam']
```

```python
# Extra steps here: Convert text to numbers
cv = CountVectorizer()
X_cv = cv.fit_transform(X)

```

```python
model = LogisticRegression()
```

```python
model.fit(X_cv, y)
```

```python
# Step 8: Try custom messages
sample =  ["Hello Christian,", "Hello Christian"]
pred = model.predict(cv.transform(sample))
if(pred[0]):
    print('Not Spam')
else:
    print('SPAM')
```
