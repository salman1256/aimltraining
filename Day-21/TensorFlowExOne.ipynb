## ** Arithmetic Operations with Tensors **

*   Add
*   Sub
*  Multi
*  Div




```python
tensor_one=tf.constant([[11,9],[2,5]])
tensor_one
```

```python
#tensor_one+15
tf.add(tensor_one,15)
```

```python
#tensor_one-5
tf.subtract(tensor_one,5)
```

```python
#tensor_one*2
tf.multiply(tensor_one,2)
```

```python
#tensor_one/2
tf.divide(tensor_one,2)
```

```python
tensor_new=tf.square(tensor_one)
tensor_new
```

```python
second_matrix=tf.constant([[8,16], [64,8.25],
                           [625,9.23],
                           [125,2.34],
                           [169,144]],dtype=tf.float32)
print(second_matrix)
```

```python
tf.sqrt(second_matrix)
```
