```python
import spacy
nlp=spacy.load("en_core_web_sm")
print('SpaCy Loadded Succcessfully')
```

```python
text='Smriti Mandhana is a cricketer from India. She have won the World cup this November 2025'
```

```python
doc=nlp(text)
for ent in doc.ents:
  print(f'\n Text: {ent.text},Label: {ent.label_ },Exp: {spacy.explain(ent.label_)} \n')
```

```python
from spacy import displacy
displacy.render(doc,style='ent',jupyter=True)
```

```python
corpus=[
    "TajMahal is located in Agra India",
    "Google was founded by Larry Page",
    "Twin Towers is located in Kuala Lumpur",
    "Discovery of India was written by Pt. Nehru"

]
```

```python
for text in corpus:
  doc=nlp(text)
  print(f'\n Text: {text}')
  for ent in doc.ents:
    print(f'{ent.text} --> {ent.label_}')
```
