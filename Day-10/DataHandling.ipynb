```python
import pandas as pd 
```

```python
df=pd.read_excel("students.xlsx")
```

```python
top_students = df[df['Marks'] > 80]
print("\n Students with Marks > 80:")
display(top_students)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read data from CSV
data = pd.read_excel("students.xlsx")

# Step 2: Filter students with marks > 80





# Step 3: Display Bar Chart
plt.figure(figsize=(10,5))
plt.bar(data["Name"], data["Marks"])

# Add chart details
plt.title("Students Marks Data")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Display chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read excel file
data = pd.read_excel("students.xlsx")

# Step 2: Filter students with marks > 80
filtered_data = data[data["Marks"] > 80]

print("Filtered Data:")
print(filtered_data)

# Step 3: Display line chart
plt.figure(figsize=(8,5))
plt.plot(filtered_data["Name"], filtered_data["Marks"], marker='o', linestyle='-', linewidth=2)

# Add chart details
plt.title("Students Scoring Above 80 Marks (Line Chart)")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True, linestyle="--", alpha=0.6)

# Display chart
plt.show()
```
