```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','April']
sales=[2000,3000,2500,3600]
```

# plt.plot(x,y,color='color name',linestyle='line_style', linewidth=value, marker='marker symbol' label='label name')

```python
plt.plot(months,sales,color='green', linestyle='--', linewidth=2, marker='*',label='monthly sales')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Report ')
plt.xlim(0,4)
plt.ylim(2000,5000)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend(loc='upper left')
plt.show()
```

# Bar Plot Example
  ## plt.bar(x,y,color='',edgecolor=' ')

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='red',label='Sales Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20K','40K','60K','80K','1M'],fontsize=12, color='green')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```
