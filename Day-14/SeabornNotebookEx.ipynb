# Seaborn Library of Matplotlib
## Seaborn is a data visualization library in Matplotlib that provides high level interface for creating interactive and more colorful plots with less code.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palette=sns.color_palette()
print(palette)
```

```python
sns.palplot(palette)
```

```python
palette_green=sns.color_palette('Greens',20)
sns.palplot(palette_green)
```

```python
data_dict={
    'building':['burz-khalifa','medrika','effil-tower','Qutub minar'],
    'category':['highest','medium','low','medium'],
    'building_height':[150,120,100,110]
    }
bar_colors=['red','green','yellow']
data=pd.DataFrame(data_dict)
sns.barplot(x='building',y='building_height',data=data, hue='category',palette=bar_colors)
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
data_iris
#data_flights=sns.load_dataset('flights')
#data_flights
```

```python
sns.barplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
plt.show

```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```
