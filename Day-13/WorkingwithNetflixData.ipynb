```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['release_year'],bins=100,color='green', edgecolor='black', alpha=0.8) 
plt.title('Netflix realeased Titles by Year ')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='blue',linestyle='--',linewidth=0.7)   
plt.tight_layout()
plt.show()  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=30,color='green', edgecolor='yellow', alpha=0.8) 
plt.title('Netflix realeased Titles by Year ')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='blue',linestyle='--',linewidth=0.7)   
plt.tight_layout()
plt.savefig('netflix_bar.pdf',dpi=300,bbox_inches='tight')
plt.show()  
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
print(movies['show_id'].count())
shows=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'], shows['release_year']], 
         bins=30,
         color=['green','blue'],
         edgecolor='magenta',
         alpha=0.8,
         label=['Movies','TV-Shows']
        ) 
plt.title('Netflix realeae by Year and Type ')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid( axis='y', color='lightblue',linestyle='--',linewidth=0.7)   
plt.tight_layout()
plt.legend(title='Movies/shows')
plt.savefig('netflix_sbowvsmovies.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

# Use netflix_data.csv dataset 
### print how many Movies and TV Shows are available on Netflix
### Display Movies and TV Shows proportion in a pie plot
### Save file as pdf



```python
df=df.dropna(subset=['type'])
```

```python
type_counts=df['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(7,7))
plt.pie(type_counts,labels=type_counts.index,  
        colors=['orange','green'],autopct='%1.2f%%', 
        startangle=140,
        shadow=True,
        explode=(0.05,0.05)
        ) 
plt.title('Netflix Movies and TV Shows Distribution')
plt.legend(title='Movies and shows')
plt.savefig('netflix_pieplot2.pdf',dpi=300)
plt.show()
```
